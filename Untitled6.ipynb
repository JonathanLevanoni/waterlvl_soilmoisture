```python
datamean = {}
for q in range(data.shape[0]):
    datamean[q] = data.iloc[q,2:].mean()
datamean = pd.Series(datamean).to_frame().reset_index()
datamean.columns = ['index', 'avg_moisture']
data = data.reset_index()
```

```python
data = data.merge(datamean, on = datamean.index)
```

```python
corr, p = spearmanr(data['SURFACE_LEVEL'], data['avg_moisture'])
print('Spearman correlation: %.3f' % corr, 'and p: %g' % p)
```

```python
fig,ax = plt.subplots(figsize=(20,10))
# make a plot
ax.plot(data.DATE,100-data['avg_moisture'],
        color="red")
# set x-axis label
ax.set_xlabel("Date", fontsize = 14)
plt.xticks(rotation=45)
# set y-axis label
ax.set_ylabel("Soil moisute",
              color="red",
              fontsize=14)
plt.xlim(['11.03.22', '08.06.22'])
#plt.ylim([0, 30])
ax2=ax.twinx()
# make a plot with different y-axis using second axis object
ax2.plot(data.DATE, data.SURFACE_LEVEL,color="blue")
ax2.set_ylabel("Surface water level",color="blue",fontsize=14)
plt.axhline(y = 0, color = 'gray', linestyle = '-')
plt.axvline(x = "11.03.22", color = 'gray')
plt.ylim([0, 40])
plt.show()
```

```python
std_error = np.std(data.avg_moisture, ddof=1) / np.sqrt(len(data.avg_moisture))
fig,ax = plt.subplots(figsize=(20,10))
# make a plot
#ax.plot(data.DATE,100-data['avg_moisture'],
#        color="red")
plt.errorbar('DATE','avg_moisture', data=data, yerr=std_error, capsize=5, color='red')
# set x-axis label
ax.set_xlabel("Date", fontsize = 14)
plt.xticks(rotation=45)
# set y-axis label
ax.set_ylabel("Soil moisture [%]",
              color="red",
              fontsize=14)
plt.xlim(['11.03.22', '08.06.22'])
plt.ylim([0, 40])
ax2=ax.twinx()
# make a plot with different y-axis using second axis object
ax2.plot(data.DATE, data.SURFACE_LEVEL,color="blue")
ax2.set_ylabel("Surface water level[cm]",color="blue",fontsize=14)
plt.axhline(y = 0, color = 'gray', linestyle = '-')
plt.axvline(x = "11.03.22", color = 'gray')
plt.ylim([0, 40])
plt.text(7.5,data.avg_moisture.min()-2, round(data.avg_moisture.min(),2), fontsize=22)
plt.text(3.5,data.avg_moisture.min()-2, 'Minimal moisture:', fontsize=22)
plt.axhline(y = data.avg_moisture.min(), color = 'purple', linestyle = ':')
plt.show()
```
